```python
classes = ['Arrachement_pelade',
           'Bouche_a_clef',
           'Comblage_de_trou_ou_Projection_d_enrobe',
           'Faiencage',
           'Grille_avaloir',
           'Longitudinale',
           'Nid_de_poule',
           'Pontage_de_fissures',
           'Raccord_de_chaussee',
           'Regard_tampon',
           'Remblaiement_de_tranchees',
           'Transversale'
]
```

```python
fig, ax = plt.subplots(figsize=(16,11))
opts = {'cmap': 'RdYlGn', 'vmin': 0, 'vmax': +1}

heatmap = ax.pcolor(matrix_normalised, **opts)
for irow in range(matrix_normalised.shape[0]):
    for icol in range(matrix_normalised.shape[1]):
        ax.text(icol+0.5, irow+0.5, '{:.2f}'.format(matrix_normalised[irow][icol]),
                ha="center", va="center", color="black")

ax.set_yticks(np.arange(0.5, matrix_normalised.shape[0], 1))
ax.set_yticklabels(classes)
ax.set_xticks(np.arange(0.5, matrix_normalised.shape[0], 1))
ax.set_xticklabels(classes, rotation=45, ha='right')
        
cbar = fig.colorbar(heatmap)
#cbar.set_label(metric_name)
fig.patch.set_facecolor('white')
```
