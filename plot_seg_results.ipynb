```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_json(path):
    with open(path, 'r') as f:
        data = json.load(f)
    return data
```

# Cracks

```python
res_bcos = read_json('data/result_seg/cracks_segmenter_bcos.json')['metric']
res_drise = read_json('data/result_seg/cracks_segmenter_drise.json')['metric']
res_rect = read_json('data/result_seg/cracks_segmenter_rectangle.json')['metric']

names = ['mIoU', 'Background', 'Arrachement pelade', 'Faiencage',
         'Nid de poule', 'Transversale', 'Longitudinale', 'Reparation']
res_bcos = [val for key, val in res_bcos.items() if 'IoU' in key]
res_drise = [val for key, val in res_drise.items() if 'IoU' in key]
res_rect = [val for key, val in res_rect.items() if 'IoU' in key]
```

```python
ind = np.arange(len(names))  # the x locations for the groups
width = 0.3  # the width of the bars

fig, ax = plt.subplots(figsize=(14,8))
rects1 = ax.barh(ind - width, res_bcos[::-1], width, label='B-cos')
rects2 = ax.barh(ind, res_drise[::-1], width, label='D-RISE')
rects3 = ax.barh(ind + width, res_rect[::-1], width, label='Rectangle')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('IoU', fontsize=16)
ax.tick_params(axis='both', which='major', labelsize=14)
ax.set_yticks(ind)
ax.set_yticklabels(names[::-1], fontsize=14)
ax.legend(fontsize=16)
plt.tight_layout()
plt.savefig('/home/thmegy/Documents/Présentations/presentations-logiroad/Plots/det_to_seg/segmenter/comparison_IoU_cracks.png')
```

# Pascal VOC

```python
res_supervised = read_json('data/result_seg/voc_segmenter_supervised.json')['metric']
res_bcos = read_json('data/result_seg/voc_segmenter_weak_bcos.json')['metric']
res_rect = read_json('data/result_seg/voc_segmenter_weak_rectangle.json')['metric']

names = ['mIoU', 'Background', 'aeroplane', 'bicycle', 'bird', 'boat', 'bottle', 'bus', 'car', 'cat', 'chair',
        'cow', 'diningtable', 'dog', 'horse', 'motorbike', 'person', 'pottedplant', 'sheep', 'sofa', 'train',
        'tvmonitor']
res_supervised = [val for key, val in res_supervised.items() if 'IoU' in key]
res_bcos = [val for key, val in res_bcos.items() if 'IoU' in key]
res_rect = [val for key, val in res_rect.items() if 'IoU' in key]
```
